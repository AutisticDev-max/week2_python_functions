```python
def read_user_file():
    filename = input("Enter the file name: ")  # ask user for file name
    try:
        f = open(filename, "r")  # try to open the file
        content = f.read()
        f.close()                # close the file
        print("File content:\n")
        print(content)           # print the file content
    except FileNotFoundError:
        print("Oops! That file does not exist.")  # handle wrong file name


read_user_file()
```
